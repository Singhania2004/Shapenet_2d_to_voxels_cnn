```python
import sys
import os
sys.path.append(os.path.abspath(".."))

from src.utils.config import load_config
from src.data.dataloader import get_dataloader
```

```python
config = load_config("../configs/config.yaml")

dataloader = get_dataloader(config)
```

```python
images, voxels = next(iter(dataloader))

print(images.shape)   # [B, 3, 128, 128]
print(voxels.shape)   # [B, 1, 32, 32, 32]
```

```python
images[0].permute(1, 2, 0).numpy()   # Convert to HWC format for visualization
voxels[0, 0].numpy()   # Get the first voxel grid for visualization
# visualizing the first image and voxel grid
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Visualize the first image
plt.figure(figsize=(6, 6))
plt.imshow(images[0].permute(1, 2, 0).numpy())
plt.title("Sample Image")
plt.axis("off")
plt.show()
# Visualize the first voxel grid
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.voxels(voxels[0, 0].numpy(), edgecolor='k')
ax.set_title("Sample Voxel Grid")
plt.show()
```
